## Проверка для умножения обычных чисел

```python
numbers = [number for number in range(1, 5001)]
times = []
retries_amount = 2000
for number in tqdm(numbers):
    start_time = monotonic()
    for ret in range(retries_amount):
        answer = fast_power(2, number)
    times.append((monotonic() - start_time) / retries_amount)
```

```python
shaping = 1e6

plt.figure(figsize=(15, 6))
plt.plot(numbers,
         [elem * shaping for elem in times],
         color="blue", 
         linestyle="-",
         label="Время работы",
         linewidth=1,
         alpha=0.6)
plt.plot(numbers,
         [1.0 * 1e-6 * np.log(number) * shaping for number in numbers],
         color='red',
         label='Асимптотика',
         linewidth=3)
plt.legend(fontsize=15)
plt.grid(True)
plt.title("График времени работы быстрого алгоритма", fontsize=20, color="green")
plt.xlabel("number", fontsize=20, color="blue")
plt.ylabel("Время, мкс", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

Можно заметить интересную вещь, что график работы на самом деле не совсем логарифмический. Связано это с тем, что умножение само по себе имеет временную сложность не O(1), а на больших степенях проделывание одного этапа умножения по порядку совпадает с временной сложностью всего возведения в степень. Если бы размер чисел не увеличивался пропорционально проделанному количеству итераций, то зависимость была бы логарифмическая

```python
numbers = [number for number in range(1, 5001)]
times = []
retries_amount = 2000
for number in tqdm(numbers):
    start_time = monotonic()
    for ret in range(retries_amount):
        answer = 2 ** number
    times.append((monotonic() - start_time) / retries_amount)

```

```python
shaping = 1e6

plt.figure(figsize=(15, 6))
plt.plot(numbers,
         [elem * shaping for elem in times],
         color="blue", 
         linestyle="-",
         label="Время работы",
         linewidth=1,
         alpha=0.6)
plt.legend(fontsize=15)
plt.grid(True)
plt.title("График времени работы быстрого алгоритма", fontsize=20, color="green")
plt.xlabel("number", fontsize=20, color="blue")
plt.ylabel("Время, мкс", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

```python
numbers = [number for number in range(1, 5001)]
times = []
retries_amount = 10
for number in tqdm(numbers):
    start_time = monotonic()
    for ret in range(retries_amount):
        answer = 1
        for i in range(number):
            answer *= 2
    times.append((monotonic() - start_time) / retries_amount)


```

```python
shaping = 1e6

plt.figure(figsize=(15, 6))
plt.plot(numbers,
         [elem * shaping for elem in times],
         color="blue", 
         linestyle="-",
         label="Время работы",
         linewidth=1,
         alpha=0.6)
plt.legend(fontsize=15)
plt.grid(True)
plt.title("График времени работы быстрого алгоритма", fontsize=20, color="green")
plt.xlabel("number", fontsize=20, color="blue")
plt.ylabel("Время, мкс", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

Этот же результат виден, если посмотреть на график зависимости для обычного возведения в степень - этот график нелинейный - нужно найти такие сущности, для которых возведение в степень не сильно наращивает число. 

Поскольку используется обычная ассоциативная операция, то можно использовать обычную сумму

## Проверка для операции суммы чисел

```python
numbers = [number for number in range(1, 10001)]
times = []
retries_amount = 1000
for number in tqdm(numbers):
    start_time = monotonic()
    for ret in range(retries_amount):
        answer = fast_power(2, number, lambda x, y: x + y, 0)
    times.append((monotonic() - start_time) / retries_amount)

```

```python
shaping = 1e6

plt.figure(figsize=(15, 6))
plt.plot(numbers,
         [elem * shaping for elem in times],
         color="blue", 
         linestyle="-",
         label="Время работы",
         linewidth=1,
         alpha=0.6)
plt.plot(numbers,
         [0.5 * 1e-6 * np.log(number) * shaping for number in numbers],
         color='red',
         label='Асимптотика',
         linewidth=3)
plt.legend(fontsize=15)
plt.grid(True)
plt.title("График времени работы быстрого алгоритма", fontsize=20, color="green")
plt.xlabel("number", fontsize=20, color="blue")
plt.ylabel("Время, мкс", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

Теперь становится гораздо лучше видна зависимость - она действительно логарифмическая
