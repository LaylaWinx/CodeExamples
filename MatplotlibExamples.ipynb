```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Simple graphs building examples

```python
x_values = [0, 5]
y_values = [0, 5]
plt.plot(x_values, y_values)
plt.show()
```

`plot()` function builds graph by connecting points to each other in the specified order. 

`x` and `y` are the arrays of coordinates. Coordinates of i-th point are ($x_i$, $y_i$). And lines in `plot-graph` are drawn between i-th and (i + 1)-th point for each i.

If there is only one array-parameter in `plot-function`, than it will be y-coordinates and x-coordinates will be (0, 1, 2, etc.)

`fmt` is the parameter for line like '{`marker`}{`line`}{`color`}'. And after this line will be drawn with this parameters. For example:

`.-r` means '.'-marker (point marker), '-'-line style (solid line), 'r'-color (red)

```python
plt.plot(x_values, y_values, ".-r")
```

Next parameter is `label`. This is the name of line for legend. Legend can be shown with `plt.legend()`

```python
plt.plot(x_values, y_values, ".-r", label="red line")
plt.legend()
```

Arguments `markerfacecolor` and `marketedgecolor` are colors for markers: first is color of inner space and second is color of edge.

Argument `markersize` is ize of markers in graph. 

```python
plt.plot(x_values,
         y_values,
         ".-b", 
         label="line",
         markerfacecolor="green", 
         markeredgecolor="red",
         markersize=20)
plt.legend()
```

One more parameter for lines is `linewidth` – this is the size of the line

```python
plt.plot(x_values,
         y_values,
         ".-b",
         label="line",
         markerfacecolor="green", 
         markeredgecolor="red",
         markersize=20,
         linewidth=2)
plt.legend()
```

Instead of using format string `fmt` in `plot()`, you can use three next parameters:

`color`: color of line

`marker`: style of marker

`linestyle`: style of the line

The last paramenter is transparency degree `alpha` – float value between 0 and 1

We can also set grid for our graph with `plt.grid(True)`.

Also, this all functions return object and if we only want to plot graph without object return – we have to write `plt.show()`

```python
plt.plot(x_values,
         y_values,
         color="blue", 
         marker='.',
         linestyle="-",
         label="line",
         markerfacecolor="green", 
         markeredgecolor="red",
         markersize=20,
         linewidth=2,
         alpha=0.6)
plt.legend()
plt.grid(True)
plt.show()
```

There are a lot of other parameters in `matplotlib.pyplot.plot()`. For example, style of plot can be changed with using `drawstyle` parameter, so that plot will drawn different type of connection between points. For example, if it is needed to plot step-plot

```python
for drawstyle in ["default", "steps-mid", "steps-pre", "steps-post"]:
    plt.plot(x_values,
             y_values,
             marker='.',
             linestyle="--",
             label=f"line {drawstyle}",
             drawstyle=drawstyle,
             linewidth=2,
             alpha=0.6)
    plt.legend()
    plt.grid(True)
plt.show()
```

Other parameters of `plot()` would not be discussed in this document. 

## Make graph more readable

Let's suppose, that we've plotted the graph. What is it? What is on x-axis, what is on y-axis, can it be clipped, what is this dependency? All these questions will be answered in this part of this document

```python
plt.plot(x_values,
         y_values,
         color="blue", 
         marker='.',
         linestyle="-",
         label="line",
         markerfacecolor="green", 
         markeredgecolor="black",
         markersize=10,
         linewidth=2,
         alpha=0.6)
plt.legend()
plt.grid(True)
plt.show()
```

Let's first set the name of this graph with `plt.title()`

Also let's set the names of x-axis and y-axis with `plt.xlabel()` and `plt.ylabel()`

```python
plt.plot(x_values,
         y_values,
         color="blue", 
         marker='.',
         linestyle="-",
         label="line",
         markerfacecolor="green", 
         markeredgecolor="black",
         markersize=10,
         linewidth=2,
         alpha=0.6)
plt.legend()
plt.grid(True)
plt.title("Dependence of x on y")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

If we think, that the tick marks on the axis are so frequent or just bad, we can set them also with use `plt.xticks()` and `plt.yticks()`

```python
plt.plot(x_values,
         y_values,
         color="blue", 
         marker='.',
         linestyle="-",
         label="line",
         markerfacecolor="green", 
         markeredgecolor="black",
         markersize=10,
         linewidth=2,
         alpha=0.6)
plt.legend()
plt.grid(True)
plt.title("Dependence of x on y")
plt.xlabel("x")
plt.ylabel("y")
plt.xticks([0, 2, 4])
plt.yticks([0, 2, 4])
plt.show()
```

There is one more parameter in this functions: `rotation`

```python
plt.plot(x_values,
         y_values,
         color="blue", 
         marker='.',
         linestyle="-",
         label="line",
         markerfacecolor="green", 
         markeredgecolor="black",
         markersize=10,
         linewidth=2,
         alpha=0.6)
plt.legend()
plt.grid(True)
plt.title("Dependence of x on y")
plt.xlabel("x")
plt.ylabel("y")
plt.xticks([0, 2, 4])
plt.yticks([0, 2, 4], rotation=90)
plt.show()
```

Also, there is possibility to make graph bigger with using `plt.figure(figsize=...)`

```python
plt.figure(figsize=(12, 8))
plt.plot(x_values,
         y_values,
         color="blue", 
         marker='.',
         linestyle="-",
         label="line",
         markerfacecolor="green", 
         markeredgecolor="black",
         markersize=10,
         linewidth=2,
         alpha=0.6)
plt.legend()
plt.grid(True)
plt.title("Dependence of x on y")
plt.xlabel("x")
plt.ylabel("y")
plt.xticks([0, 2, 4])
plt.yticks([0, 2, 4], rotation=90)
plt.show()
```

But when we set big size for this plot, text sizes don't become bigger. And we have to make text-size bigger.

Text of labels can be set with parameter `fontsize` in `plt.xlabel()` and `plt.ylabel()`

Ticks size can be changed with parameter `fontsize` in `plt.xticks()` and `plt.yticks()`

Size of title with parameter `fontsize` in `plt.title()`

Also, we can set the color for these both texts with parameter `color`

The size of legend can be set with `fontsize` in `plt.legend()`

```python
plt.figure(figsize=(12, 8))
plt.plot(x_values,
         y_values,
         color="blue", 
         marker='.',
         linestyle="-",
         label="line",
         markerfacecolor="green", 
         markeredgecolor="black",
         markersize=10,
         linewidth=2,
         alpha=0.6)
plt.legend(fontsize=15)
plt.grid(True)
plt.title("Dependence of x on y", fontsize=20, color="green")
plt.xlabel("x", fontsize=20, color="blue")
plt.ylabel("y", fontsize=20, rotation=0, color="blue")
plt.xticks([0, 2, 4], fontsize=20, color="red")
plt.yticks([0, 2, 4], fontsize=20, color="red")
plt.show()
```

And now we are ready to make another graph and compare, what was before setting of parameters and after

```python
import numpy as np

x_values = np.array([10 * i for i in range(100)])
y_values = np.sin(x_values / 180 * np.pi)

# -----BEFORE------
plt.plot(x_values, y_values)
plt.show()

# -----AFTER-------
plt.figure(figsize=(15, 6))
plt.plot(x_values,
         y_values,
         color="blue", 
         marker='.',
         linestyle="-",
         label="sine line",
         markerfacecolor="green", 
         markeredgecolor="black",
         markersize=10,
         linewidth=2,
         alpha=0.6)
plt.legend(fontsize=15)
plt.grid(True)
plt.title("Sine graph in degrees", fontsize=20, color="green")
plt.xlabel("x, deg", fontsize=20, color="blue")
plt.ylabel("sin(x)", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

As we can see, we make this graph more readable and more beautiful

## Histograms

`plt.hist()` is a histogram of the set of points

```python
values = np.random.normal(loc=0.0, scale=1.0, size=100_000)
```

```python
plt.figure(figsize=(12, 6))
plt.hist(values, label="normal distribution")
plt.grid(True)
plt.legend(fontsize=15)
plt.title("Histogram of normal distribution", fontsize=20, color="green")
plt.xlabel("Value", fontsize=20, color="blue")
plt.ylabel("Amount", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

We can plot histogram in values range, so it just drops values, that are not in range. It can be made with parameter `range`

```python
plt.figure(figsize=(12, 6))
plt.hist(values, label="normal distribution", range=(-2, 2))
plt.grid(True)
plt.legend(fontsize=15)
plt.title("Histogram of normal distribution", fontsize=20, color="green")
plt.xlabel("Value", fontsize=20, color="blue")
plt.ylabel("Amount", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

As we can see, on `y-axis` there is an amount of values for every bin. If we want to see the part of values in this bin (density of values) we can set parameter `density=True` in `plt.hist()`

```python
plt.figure(figsize=(12, 6))
plt.hist(values, label="normal distribution", density=True, range=(-3, 3))
plt.grid(True)
plt.legend(fontsize=15)
plt.title("Histogram of normal distribution", fontsize=20, color="green")
plt.xlabel("Value", fontsize=20, color="blue")
plt.ylabel("Amount", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

This is the plot of density of values. It is the estimation for `pdf`. We can make the estimation of `cdf` – this is just cumulative graph of this density. It can be made by set `cumulative=True`

```python
plt.figure(figsize=(12, 6))
plt.hist(values, label="normal distribution", density=True, range=(-3, 3), cumulative=True)
plt.grid(True)
plt.legend(fontsize=15)
plt.title("Histogram of normal distribution", fontsize=20, color="green")
plt.xlabel("Value", fontsize=20, color="blue")
plt.ylabel("Amount", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

It's hard to see anything here. That's why it's also important to know, how to make more bins in histplot. It can be made by set parameter `bins`

```python
plt.figure(figsize=(12, 6))
plt.hist(values, label="normal distribution", density=True, range=(-3, 3), cumulative=True, bins=300)
plt.grid(True)
plt.legend(fontsize=15)
plt.title("Histogram of normal distribution", fontsize=20, color="green")
plt.xlabel("Value", fontsize=20, color="blue")
plt.ylabel("CDF", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(values, label="normal distribution", density=True, range=(-3, 3), bins=300)
plt.grid(True)
plt.legend(fontsize=15)
plt.title("Histogram of normal distribution", fontsize=20, color="green")
plt.xlabel("Value", fontsize=20, color="blue")
plt.ylabel("PDF", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

## Scatter plot

Scatter plot is used to plot points on the graph. It has 2 main parameters: `x`: array of x-coordinates, and `y`: array of y-coordinates.

```python
x_values = [0, 5, 10, 5]
y_values = [0, 6, 0, 2]
plt.scatter(x_values, y_values)
```

It may become, that you need to change size of points, because there are a lot of points. For example, in this scatter-plot

```python
x_values = np.random.normal(loc=0.0, scale=0.3, size=100_000)
y_values = np.random.normal(loc=0.0, scale=0.3, size=100_000)
```

```python
plt.figure(figsize=(12, 9))
plt.scatter(x_values, y_values, label="2d-normal-distributed points")
plt.grid(True)
plt.legend(fontsize=15)
plt.title("Points from normal-2D", fontsize=20, color="green")
plt.xlabel("x_value", fontsize=20, color="blue")
plt.ylabel("y-value", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

This just looks like a big cloud. If we change the size of every point with parameter `s`, we will get:

```python
plt.figure(figsize=(12, 9))
plt.scatter(x_values, y_values, s=0.1, label="2d-normal-distributed points")
plt.grid(True)
plt.legend(fontsize=15)
plt.title("Points from normal-2D", fontsize=20, color="green")
plt.xlabel("x_value", fontsize=20, color="blue")
plt.ylabel("y-value", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

As we see, this graph has more vosible information about points and we see that visible part of cloud is lower, because probability is higher

We also can set the color for every point with parameter `c`. Default sequence for different colors is taken from `image.cmap`

```python
plt.figure(figsize=(12, 9))
plt.scatter(x_values,
            y_values,
            s=0.1,
            c="r",
            label="2d-normal-distributed points")
plt.grid(True)
plt.legend(fontsize=15)
plt.title("Points from normal-2D", fontsize=20, color="green")
plt.xlabel("x_value", fontsize=20, color="blue")
plt.ylabel("y-value", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

For bigger visible effect, points can be transparent. For this you have to set parameter `alpha`

```python
plt.figure(figsize=(12, 9))
plt.scatter(x_values,
            y_values,
            s=0.1,
            c="r",
            alpha=0.2,
            label="2d-normal-distributed points")
plt.grid(True)
plt.legend(fontsize=15)
plt.title("Points from normal-2D", fontsize=20, color="green")
plt.xlabel("x_value", fontsize=20, color="blue")
plt.ylabel("y-value", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

And now we can combine these graphs to solve any problem. For example, we need to get dependence from noisy-data

```python
x_values = np.array([0.1 * i for i in range(1000)])
y_values = 0.1 * x_values ** 2 + x_values + 100 * np.sin(x_values)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(x_values,
         y_values,
         color="blue",
         marker='.',
         linestyle="-",
         label="sine line",
         markerfacecolor="green",
         markeredgecolor="black",
         markersize=10,
         linewidth=2,
         alpha=0.6)
plt.legend(fontsize=15)
plt.grid(True)
plt.title("Dependence", fontsize=20, color="green")
plt.xlabel("x", fontsize=20, color="blue")
plt.ylabel("f(x)", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

Now, generate points with little noise from this function

```python
noisy_y_values = y_values + np.random.normal(loc=0.0, scale=40, size=len(y_values))
```

```python
plt.figure(figsize=(15, 6))
plt.plot(x_values,
         y_values,
         color="blue",
         marker='.',
         linestyle="-",
         label="true function line",
         markerfacecolor="green",
         markeredgecolor="black",
         markersize=10,
         linewidth=2,
         alpha=0.6)
plt.scatter(x_values,
            noisy_y_values,
            s=5,
            c="r",
            label="Noisy points")
plt.legend(fontsize=15)
plt.grid(True)
plt.title("Dependence", fontsize=20, color="green")
plt.xlabel("x", fontsize=20, color="blue")
plt.ylabel("f(x)", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.scatter(x_values,
            noisy_y_values,
            s=5,
            c="r",
            label="Noisy points")
plt.legend(fontsize=15)
plt.grid(True)
plt.title("Dependence", fontsize=20, color="green")
plt.xlabel("x", fontsize=20, color="blue")
plt.ylabel("f(x)", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```

Now let's try to make some approximations and get distribution of errors

```python
from sklearn.linear_model import LinearRegression

plt.figure(figsize=(15, 6))
plt.scatter(x_values,
            noisy_y_values,
            s=5,
            c="r",
            label="Noisy points")


# Simple linear function
reg = LinearRegression().fit(x_values.reshape(-1, 1), y_values)
plt.plot(x_values,
         reg.predict(x_values.reshape(-1, 1)),
         color="blue",
         linestyle="-",
         label="Linear estimation",
         linewidth=2)
linear_errors = noisy_y_values - reg.predict(x_values.reshape(-1, 1))

# Quadratic function
reg = LinearRegression().fit([[x, x ** 2] for x in x_values], y_values)
plt.plot(x_values,
         reg.predict([[x, x ** 2] for x in x_values]),
         color="green",
         linestyle="-",
         label="Quadratic estimation",
         linewidth=2)
quadratic_errors = noisy_y_values - reg.predict([[x, x ** 2] for x in x_values])

# Estimation with sin(x), x, x^2
reg = LinearRegression().fit([[x, x ** 2, np.sin(x)] for x in x_values], y_values)
plt.plot(x_values,
         reg.predict([[x, x ** 2, np.sin(x)] for x in x_values]),
         color="black",
         linestyle="-",
         label="Right estimation",
         linewidth=2)
final_errors = noisy_y_values - reg.predict([[x, x ** 2, np.sin(x)] for x in x_values])


plt.legend(fontsize=15)
plt.grid(True)
plt.title("Dependence", fontsize=20, color="green")
plt.xlabel("x", fontsize=20, color="blue")
plt.ylabel("f(x)", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
```

```python
plt.figure(figsize=(12, 6))
plt.hist(linear_errors, label="Linear model errors distribution", density=True, bins=30, alpha=0.5)
plt.hist(quadratic_errors, label="Quadratic model errors distribution", density=True, bins=30, alpha=0.5)
plt.hist(final_errors, label="Final model errors distribution", density=True, bins=30, alpha=0.5)
plt.grid(True)
plt.legend(fontsize=15)
plt.title("Histogram of normal distribution", fontsize=20, color="green")
plt.xlabel("Value", fontsize=20, color="blue")
plt.ylabel("PDF", fontsize=20, rotation=0, color="blue")
plt.xticks(fontsize=20, color="red")
plt.yticks(fontsize=20, color="red")
plt.show()
```
